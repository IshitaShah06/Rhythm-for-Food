```python
max_week = int(df.week.max())
min_week = int(df.week.min())
max_val = int(df.value.max())

def plot_year_circle(df, name, years, max_week=max_week, min_week=min_week, max_val=max_val, fig_size=8):
    data = df[(df.id == name) & (df.year.isin(years))]
    weeks = data.week.values / max_week
    values = data.value.values / max_val
    color_map = dict(zip(years, sns.color_palette('dark', n_colors=len(years))))
    
    theta = 2 * np.pi * weeks
    area = 200 * values**2
    colors = data.year.apply(color_map.get).values

    fig = plt.figure(figsize=(fig_size, fig_size))
    ax = fig.add_subplot(111, projection='polar')
    ax.scatter(theta, values, c=colors, s=area, cmap='hsv', alpha=0.75)
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    angles = [360./13.*i for i in range(min_week, int(max_week/4)+1)]

    labels = [i*4 for i in range(min_week, int(max_week/4)+1)]
    ax.set_thetagrids(angles, labels)
    ax.set_title(f"Year Circle for {name} in {years}".title(), fontsize=16)

    marks = []
    labels = []
    for year, color in color_map.items():
        marks.append(mpatches.Circle((0,0), radius=1, fc=color))
        labels.append(year)
    plt.legend(marks, years)
    
    ax.legend()
    plt.draw()

def plot_weekly_avg(data, name, accented=None, accented_size=1.7, size=0.7):
    subset = df[df.id.isin(data)]
    weeklyAvg = subset.groupby(['year', 'id']).value.mean().to_frame().reset_index()
    line_size = size
    if accented:
        line_size = weeklyAvg.id.map(lambda x: accented_size if x == accented else size)
    (
        ggplot(weeklyAvg, aes('year', 'value', color='id'))
        + geom_point()
        + geom_line(size=line_size)
        + ggtitle('Weekly Average for ' + name)
    ).draw()
    #weeklyAverage.pivot('year', 'id', 'value').plot(figsize=(12,8)) ## plot with pandas

def get_names_by_class(food_class):
    return [name for name, klass in class_map.items() if klass == food_class]

def plot_food_set(food_set, years):
    foodtrend = df[df.id.isin(food_set) & df.year.isin(years)]

    (
        ggplot(foodtrend, aes('week', 'value', color='id'))
         + geom_point()
    ).draw()
```
